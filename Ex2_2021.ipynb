DataFrame Cleaning:
- Use `pandas.DataFrame.shape` to determine:

- Rename the variables **"Country name", "Regional indicator", "Ladder score", "Explained by: Log GDP per capita", "Explained by: Social support", "Explained by: Healthy life expectancy", "Explained by: Freedom to make life choices", "Explained by: Generosity", "Explained by: Perceptions of corruption"** to **“Country”, "Region", “Happiness”, “Economy”, “Family”, “Health”, “Freedom”, “Trust”** respectively using `pandas.DataFrame.rename()`.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# read csv
df = pd.read_csv("2020 World Happiness Report.csv")

# display DataFrame
display(df)

# Print DataFrame shape
df.shape
```

```python
# drop columns
df.drop(['Standard error of ladder score',
         'upperwhisker',
         'lowerwhisker',
         'Logged GDP per capita',
         'Social support',
         'Healthy life expectancy',
         'Freedom to make life choices',
         'Generosity',
         'Perceptions of corruption',
         'Ladder score in Dystopia',
         'Dystopia + residual'], axis = 1, inplace=True)
# rename columns
df.rename(columns={"Country name": "Country", 
                    "Regional indicator": "Region", 
                    "Ladder score": "Happiness", 
                    "Explained by: Log GDP per capita": "Economy", 
                    "Explained by: Social support": "Family", 
                    "Explained by: Healthy life expectancy": "Health", 
                    "Explained by: Freedom to make life choices": "Freedom", 
                    "Explained by: Generosity": "Generosity", 
                    "Explained by: Perceptions of corruption": "Trust"}, inplace=True)
# display DataFrame
display(df)
```

```python
# print DataFrame info
df.info()
```

```python
# convert Region to categorical
df = df.astype({"Region":'category'})

# print Region Categories
print(df['Region'].cat.categories)
```

```python
# print min, max, mean Happiness
print(df["Happiness"].min(), df["Happiness"].max(), df["Happiness"].mean())
```
